# Output Data 1 create a resultant CSV/excel file with following columns:
- Order ID
- AWB number
- Total Weight as per X(KG)
- Weight slab as per X(KG)
- Total weight as per Courier Company(KG)
- Weight slab charged by Courier Company(KG)
- Delivery Zone as per X
- Delivery Zone charged by Courier company
- Expected charge as per X(Rs.)
- Charges billed by courier company(Rs.)
- Difference Between expected charges and billed charges(Rs.)

```python
#3
# merge data X company order and company master SKU data on SKU

# merge df1 and df3

df = pd.merge(df1, df3, on= 'SKU')

# created new column for total weight base on gross weight and quantity
#  to calculate total weight of each order 

df['Total Weight X'] = df['Order Qty'] * df['Weight (g)']

# drop the not important columns like SKU, Order Qty, Weight (g)

df = df.drop(columns= ['SKU', 'Order Qty', 'Weight (g)'])

#4
# merge the df4(Courier Company - Invoice.xlsx) to the 'df' frame on Order ID

df = df.merge(df4, on='Order ID')

# calculated weight base on courier company data
df['Total Weight C'] = df['Billing Amount (Rs.)']/df['Charged Weight']

# merge the df5(Courier Company - Rates.xlsx) baseon on 'Zone'

df = df.merge(df5, on='Zone')
```

```python
#5
# created calculated conditional column base on shipment charges
# c - charges

for i in df['Type of Shipment']:
    if i == 'Forward charges':
        df['c'] = df['Forward Fixed Charge'] + df['Forward Additional Weight Slab Charge']
    else:
        df['c'] = df['Forward Fixed Charge'] + df['Forward Additional Weight Slab Charge']+df['RTO Fixed Charge'] + df['RTO Additional Weight Slab Charge']
df
```

```python
# drop the 'Type of Shipment', 'Forward Fixed Charge', 'Forward Additional Weight Slab Charge','RTO Fixed Charge','RTO Additional Weight Slab Charge'

df = df.drop(columns=['Type of Shipment', 'Forward Fixed Charge', 'Forward Additional Weight Slab Charge','RTO Fixed Charge','RTO Additional Weight Slab Charge'])
df
```
